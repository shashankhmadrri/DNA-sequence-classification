```python
import os
import random
import warnings
import zipfile
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, classification_report, roc_curve
)
from sklearn.utils.class_weight import compute_class_weight
from sklearn.model_selection import train_test_split

warnings.filterwarnings("ignore")
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"


# 1. Reproducibility

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)


# 2. Config

MAX_LEN = 251            # dataset sequence length
BATCH_SIZE = 128
EPOCHS = 8
VALIDATION_SPLIT = 0.1
LEARNING_RATE = 1e-3


zip_candidates = [
    Path("/content/promoter_dataset.zip"),
    Path("promoter_dataset.zip")
]

extract_candidates = [
    Path("/content/promoter_dataset"),
    Path("promoter_dataset")
]

dataset_root = None

# Case 1: extracted folder already exists
for p in extract_candidates:
    if p.exists() and p.is_dir():
        dataset_root = p
        break

# Case 2: zip exists -> extract
if dataset_root is None:
    zip_path = None
    for z in zip_candidates:
        if z.exists() and z.is_file():
            zip_path = z
            break

    if zip_path is None:
        raise FileNotFoundError(
            "Could not find promoter_dataset.zip or extracted promoter_dataset folder in Colab. "
            "Upload promoter_dataset.zip first."
        )

    dataset_root = Path("/content/promoter_dataset")
    dataset_root.mkdir(parents=True, exist_ok=True)

    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(dataset_root)

    # Some zip files contain train/ and test/ directly at root,
    # others may contain an extra top-level folder.
    if not (dataset_root / "train").exists():
        subdirs = [d for d in dataset_root.iterdir() if d.is_dir()]
        for d in subdirs:
            if (d / "train").exists() and (d / "test").exists():
                dataset_root = d
                break

print(f"Dataset root: {dataset_root}")

train_dir = dataset_root / "train"
test_dir = dataset_root / "test"

if not train_dir.exists() or not test_dir.exists():
    raise FileNotFoundError(
        f"Expected train/ and test/ inside {dataset_root}, but they were not found."
    )

print("Train dir:", train_dir)
print("Test dir: ", test_dir)


# 4. Load sequences from folders

def read_sequences_from_split(split_dir):
    sequences = []
    labels = []
    class_map = {"negative": 0, "positive": 1}

    for class_name, label in class_map.items():
        class_dir = split_dir / class_name
        if not class_dir.exists():
            raise FileNotFoundError(f"Missing folder: {class_dir}")

        for file_path in sorted(class_dir.glob("*")):
            if file_path.is_file():
                seq = file_path.read_text().strip().upper()
                sequences.append(seq)
                labels.append(label)

    return sequences, np.array(labels, dtype=np.int32)

train_sequences, y_train = read_sequences_from_split(train_dir)
test_sequences, y_test = read_sequences_from_split(test_dir)

print(f"\nTrain samples: {len(train_sequences)}")
print(f"Test samples:  {len(test_sequences)}")
print(f"Example sequence: {train_sequences[0][:60]}...")


# 5. DNA one-hot encoding

# A, C, G, T -> 4 channels

base_to_vec = {
    "A": [1, 0, 0, 0],
    "C": [0, 1, 0, 0],
    "G": [0, 0, 1, 0],
    "T": [0, 0, 0, 1]
}

def one_hot_encode_sequence(seq, max_len=251):
    seq = seq.upper()
    arr = np.zeros((max_len, 4), dtype=np.float32)
    for i, ch in enumerate(seq[:max_len]):
        if ch in base_to_vec:
            arr[i] = base_to_vec[ch]
        # unknown characters remain [0,0,0,0]
    return arr

def encode_sequences(seq_list, max_len=251):
    return np.stack([one_hot_encode_sequence(seq, max_len) for seq in seq_list])

print("\nEncoding sequences...")
X_train = encode_sequences(train_sequences, MAX_LEN)
X_test = encode_sequences(test_sequences, MAX_LEN)

print("X_train shape:", X_train.shape)
print("X_test shape: ", X_test.shape)


# 6. Train/validation split

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train,
    test_size=VALIDATION_SPLIT,
    stratify=y_train,
    random_state=SEED
)

print("\nAfter validation split:")
print("Train:", X_tr.shape, y_tr.shape)
print("Val:  ", X_val.shape, y_val.shape)


# 7. Class weights

classes = np.unique(y_tr)
weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_tr)
class_weights = {int(c): float(w) for c, w in zip(classes, weights)}
print("\nClass weights:", class_weights)


# 8. Model builders

def build_cnn(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=input_shape),
        tf.keras.layers.Conv1D(64, kernel_size=8, activation="relu", padding="same"),
        tf.keras.layers.MaxPooling1D(pool_size=2),
        tf.keras.layers.Conv1D(128, kernel_size=8, activation="relu", padding="same"),
        tf.keras.layers.MaxPooling1D(pool_size=2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])
    return model

def build_lstm(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=input_shape),
        tf.keras.layers.LSTM(64, return_sequences=False),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(64, activation="relu"),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])
    return model

def build_cnn_bilstm(input_shape):
    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=input_shape),
        tf.keras.layers.Conv1D(64, kernel_size=8, activation="relu", padding="same"),
        tf.keras.layers.MaxPooling1D(pool_size=2),
        tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences=False)),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(128, activation="relu"),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(1, activation="sigmoid")
    ])
    return model


# 9. Compile, train, evaluate

def compile_model(model):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="auc")
        ]
    )
    return model

callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_auc",
        mode="max",
        patience=3,
        restore_best_weights=True
    )
]

def evaluate_model(model, X, y_true, model_name="Model"):
    y_prob = model.predict(X, batch_size=BATCH_SIZE, verbose=0).ravel()
    y_pred = (y_prob >= 0.5).astype(int)

    metrics = {
        "Model": model_name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, zero_division=0),
        "Recall": recall_score(y_true, y_pred, zero_division=0),
        "F1": f1_score(y_true, y_pred, zero_division=0),
        "ROC_AUC": roc_auc_score(y_true, y_prob)
    }
    cm = confusion_matrix(y_true, y_pred)
    return metrics, cm, y_prob, y_pred

def plot_training_history(history, title):
    hist = pd.DataFrame(history.history)

    plt.figure(figsize=(7, 4))
    plt.plot(hist["loss"], label="train_loss")
    plt.plot(hist["val_loss"], label="val_loss")
    plt.title(f"{title} - Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()

    if "auc" in hist.columns and "val_auc" in hist.columns:
        plt.figure(figsize=(7, 4))
        plt.plot(hist["auc"], label="train_auc")
        plt.plot(hist["val_auc"], label="val_auc")
        plt.title(f"{title} - AUC")
        plt.xlabel("Epoch")
        plt.ylabel("AUC")
        plt.legend()
        plt.show()

def plot_confusion_matrix(cm, title):
    plt.figure(figsize=(5, 4))
    plt.imshow(cm, interpolation="nearest")
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(2)
    plt.xticks(tick_marks, ["Non-Promoter", "Promoter"])
    plt.yticks(tick_marks, ["Non-Promoter", "Promoter"])
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")

    thresh = cm.max() / 2.0
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(
                j, i, str(cm[i, j]),
                horizontalalignment="center",
                color="white" if cm[i, j] > thresh else "black"
            )
    plt.tight_layout()
    plt.show()

def plot_roc_curve(y_true, y_prob, title):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    auc_score = roc_auc_score(y_true, y_prob)
    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f"AUC = {auc_score:.4f}")
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right")
    plt.show()

models = {
    "CNN": build_cnn((MAX_LEN, 4)),
    "LSTM": build_lstm((MAX_LEN, 4)),
    "CNN_BiLSTM": build_cnn_bilstm((MAX_LEN, 4))
}

results = []
saved_objects = {}

for model_name, model in models.items():
    print("\n" + "=" * 70)
    print(f"Training {model_name}")
    print("=" * 70)

    model = compile_model(model)
    model.summary()

    history = model.fit(
        X_tr, y_tr,
        validation_data=(X_val, y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        class_weight=class_weights,
        callbacks=callbacks,
        verbose=1
    )

    plot_training_history(history, model_name)

    metrics, cm, y_prob, y_pred = evaluate_model(model, X_test, y_test, model_name)
    results.append(metrics)
    saved_objects[model_name] = {
        "model": model,
        "history": history,
        "cm": cm,
        "y_prob": y_prob,
        "y_pred": y_pred
    }

    print(f"\nTest Classification Report for {model_name}:")
    print(classification_report(y_test, y_pred, target_names=["Non-Promoter", "Promoter"]))
    print("Confusion Matrix:\n", cm)
    print("Metrics:", metrics)

    plot_confusion_matrix(cm, f"{model_name} - Confusion Matrix")
    plot_roc_curve(y_test, y_prob, f"{model_name} - ROC Curve")


# 10. Compare models

results_df = pd.DataFrame(results).sort_values(by="ROC_AUC", ascending=False).reset_index(drop=True)

print("\n" + "=" * 70)
print("FINAL MODEL COMPARISON")
print("=" * 70)
print(results_df)

best_model_name = results_df.loc[0, "Model"]
best_model = saved_objects[best_model_name]["model"]

print(f"\nBest model based on ROC_AUC: {best_model_name}")


# 11. Save best model

best_model.save("best_promoter_classifier.keras")
print("\nSaved best model as: best_promoter_classifier.keras")


# 12. Prediction function

def predict_sequence(seq, model, max_len=251):
    x = one_hot_encode_sequence(seq, max_len=max_len)
    x = np.expand_dims(x, axis=0)
    prob = float(model.predict(x, verbose=0)[0][0])
    label = "Promoter" if prob >= 0.5 else "Non-Promoter"
    return {"sequence": seq, "predicted_label": label, "probability_promoter": prob}

# Example prediction using one test sample
example_seq = test_sequences[0]
example_result = predict_sequence(example_seq, best_model, max_len=MAX_LEN)

print("\nExample prediction with best model:")
print(example_result)
```
