```python
import os
import re
import json
import random
import warnings
import zipfile
from pathlib import Path
from itertools import product
from collections import Counter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    confusion_matrix,
    classification_report,
    roc_curve,
    precision_recall_curve,
)

warnings.filterwarnings("ignore")
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"

# ============================================================
# 1. REPRODUCIBILITY
# ============================================================

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ============================================================
# 2. CONFIGURATION
# ============================================================

MAX_LEN = 251
BATCH_SIZE = 128
TUNING_EPOCHS = 4
FINAL_EPOCHS = 12
VALIDATION_SPLIT = 0.10
BASE_LEARNING_RATE = 1e-3
OUTPUT_DIR = Path("promoter_final_outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)


DANQ_REFERENCE_ROC_AUC = 0.909
DANQ_REFERENCE_NOTE = (
    "Configurable DanQ reference. Replace with the exact DanQ target ROC-AUC "

)

# ============================================================
# 3. DATASET DISCOVERY AND EXTRACTION
# ============================================================

zip_candidates = [
    Path("/content/drive/MyDrive/promoter_dataset.zip"),
    Path("/content/promoter_dataset.zip"),
    Path("promoter_dataset.zip"),
]

extract_candidates = [
    Path("/content/promoter_dataset"),
    Path("/content/drive/MyDrive/promoter_dataset"),
    Path("promoter_dataset"),
]


def find_dataset_root():
    for p in extract_candidates:
        if p.exists() and p.is_dir():
            if (p / "train").exists() and (p / "test").exists():
                return p
            for sub in p.iterdir():
                if sub.is_dir() and (sub / "train").exists() and (sub / "test").exists():
                    return sub

    zip_path = None
    for z in zip_candidates:
        if z.exists() and z.is_file():
            zip_path = z
            break

    if zip_path is None:
        raise FileNotFoundError(
            "Could not find promoter_dataset.zip or extracted promoter_dataset folder.\n"
            "Put promoter_dataset.zip in the current folder, /content/, or /content/drive/MyDrive/."
        )

    dataset_root = Path("/content/promoter_dataset") if Path("/content").exists() else Path("promoter_dataset")
    dataset_root.mkdir(parents=True, exist_ok=True)

    print(f"Extracting {zip_path} to {dataset_root} ...")
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
        zip_ref.extractall(dataset_root)

    if (dataset_root / "train").exists() and (dataset_root / "test").exists():
        return dataset_root

    for sub in dataset_root.iterdir():
        if sub.is_dir() and (sub / "train").exists() and (sub / "test").exists():
            return sub

    raise FileNotFoundError(
        f"After extraction, train/ and test/ folders were not found inside {dataset_root}."
    )


dataset_root = find_dataset_root()
train_dir = dataset_root / "train"
test_dir = dataset_root / "test"

print("Dataset root:", dataset_root)
print("Train dir:", train_dir)
print("Test dir:", test_dir)

# ============================================================
# 4. LOAD SEQUENCES
# ============================================================


def clean_sequence(seq):
    seq = seq.strip().upper()
    seq = re.sub(r"[^ACGTN]", "N", seq)
    return seq


def read_sequences_from_split(split_dir):
    sequences = []
    labels = []
    class_map = {"negative": 0, "positive": 1}

    for class_name, label in class_map.items():
        class_dir = split_dir / class_name
        if not class_dir.exists():
            raise FileNotFoundError(f"Missing folder: {class_dir}")

        for file_path in sorted(class_dir.glob("*")):
            if file_path.is_file():
                seq = clean_sequence(file_path.read_text())
                if len(seq) > 0:
                    sequences.append(seq)
                    labels.append(label)

    return sequences, np.array(labels, dtype=np.int32)


train_sequences, y_train = read_sequences_from_split(train_dir)
test_sequences, y_test = read_sequences_from_split(test_dir)

print(f"\nTrain samples: {len(train_sequences)}")
print(f"Test samples:  {len(test_sequences)}")
print("Train labels:", Counter(y_train))
print("Test labels: ", Counter(y_test))
print(f"Example sequence: {train_sequences[0][:80]}...")

# ============================================================
# 5. ONE-HOT ENCODING
# ============================================================

base_to_vec = {
    "A": [1, 0, 0, 0],
    "C": [0, 1, 0, 0],
    "G": [0, 0, 1, 0],
    "T": [0, 0, 0, 1],
    "N": [0, 0, 0, 0],
}


def one_hot_encode_sequence(seq, max_len=MAX_LEN):
    arr = np.zeros((max_len, 4), dtype=np.float32)
    seq = clean_sequence(seq)
    for i, ch in enumerate(seq[:max_len]):
        arr[i] = base_to_vec.get(ch, [0, 0, 0, 0])
    return arr


def encode_sequences(seq_list, max_len=MAX_LEN):
    return np.stack([one_hot_encode_sequence(seq, max_len) for seq in seq_list])


print("\nEncoding sequences...")
X_train = encode_sequences(train_sequences, MAX_LEN)
X_test = encode_sequences(test_sequences, MAX_LEN)

X_tr, X_val, y_tr, y_val, seq_tr, seq_val = train_test_split(
    X_train,
    y_train,
    train_sequences,
    test_size=VALIDATION_SPLIT,
    stratify=y_train,
    random_state=SEED,
)

classes = np.unique(y_tr)
weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_tr)
class_weights = {int(c): float(w) for c, w in zip(classes, weights)}

print("X_train:", X_train.shape)
print("X_val:  ", X_val.shape)
print("X_test: ", X_test.shape)
print("Class weights:", class_weights)

# ============================================================
# 6. MODEL BUILDERS
# ============================================================


def build_cnn(input_shape, filters=64, kernel_size=8, dense_units=128, dropout=0.50, learning_rate=1e-3):
    inputs = tf.keras.layers.Input(shape=input_shape)
    x = tf.keras.layers.Conv1D(filters, kernel_size=kernel_size, padding="same", activation="relu", name="motif_conv_1")(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling1D(pool_size=2)(x)
    x = tf.keras.layers.Conv1D(filters * 2, kernel_size=kernel_size, padding="same", activation="relu", name="motif_conv_2")(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling1D(pool_size=2)(x)
    x = tf.keras.layers.GlobalMaxPooling1D(name="global_motif_pooling")(x)
    x = tf.keras.layers.Dense(dense_units, activation="relu")(x)
    x = tf.keras.layers.Dropout(dropout)(x)
    outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)

    model = tf.keras.Model(inputs, outputs, name=f"CNN_f{filters}_k{kernel_size}_d{dense_units}")
    compile_model(model, learning_rate)
    return model


def build_hybrid_cnn_bilstm(input_shape, filters=64, kernel_size=8, lstm_units=64, dense_units=128, dropout=0.50, learning_rate=1e-3):
    inputs = tf.keras.layers.Input(shape=input_shape)
    x = tf.keras.layers.Conv1D(filters, kernel_size=kernel_size, padding="same", activation="relu", name="motif_conv")(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPooling1D(pool_size=2)(x)

    # DanQ-style idea: CNN captures local motifs; BiLSTM captures longer-range motif dependencies.
    x = tf.keras.layers.Bidirectional(
        tf.keras.layers.LSTM(lstm_units, return_sequences=False),
        name="long_range_dependency_bilstm",
    )(x)

    x = tf.keras.layers.Dropout(dropout)(x)
    x = tf.keras.layers.Dense(dense_units, activation="relu")(x)
    x = tf.keras.layers.Dropout(dropout / 2)(x)
    outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)

    model = tf.keras.Model(inputs, outputs, name=f"Hybrid_CNN_BiLSTM_f{filters}_k{kernel_size}_l{lstm_units}")
    compile_model(model, learning_rate)
    return model


def compile_model(model, learning_rate):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(name="precision"),
            tf.keras.metrics.Recall(name="recall"),
            tf.keras.metrics.AUC(name="roc_auc", curve="ROC"),
            tf.keras.metrics.AUC(name="pr_auc", curve="PR"),
        ],
    )

# ============================================================
# 7. PLOTTING FUNCTIONS - HIGH RESOLUTION
# ============================================================


def savefig(name):
    path = OUTPUT_DIR / name
    plt.tight_layout()
    plt.savefig(path, dpi=300, bbox_inches="tight")
    plt.show()
    print("Saved:", path)


def plot_training_history(history, model_name):
    hist = pd.DataFrame(history.history)

    plt.figure(figsize=(8, 5))
    plt.plot(hist["loss"], label="Training Loss", linewidth=2)
    plt.plot(hist["val_loss"], label="Validation Loss", linewidth=2)
    plt.title(f"{model_name}: Training and Validation Loss", fontsize=14)
    plt.xlabel("Epoch", fontsize=12)
    plt.ylabel("Binary Cross-Entropy Loss", fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(alpha=0.3)
    savefig(f"{model_name}_loss.png")

    plt.figure(figsize=(8, 5))
    plt.plot(hist["roc_auc"], label="Training ROC-AUC", linewidth=2)
    plt.plot(hist["val_roc_auc"], label="Validation ROC-AUC", linewidth=2)
    plt.title(f"{model_name}: ROC-AUC During Training", fontsize=14)
    plt.xlabel("Epoch", fontsize=12)
    plt.ylabel("ROC-AUC", fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(alpha=0.3)
    savefig(f"{model_name}_training_roc_auc.png")

    plt.figure(figsize=(8, 5))
    plt.plot(hist["pr_auc"], label="Training PR-AUC", linewidth=2)
    plt.plot(hist["val_pr_auc"], label="Validation PR-AUC", linewidth=2)
    plt.title(f"{model_name}: PR-AUC During Training", fontsize=14)
    plt.xlabel("Epoch", fontsize=12)
    plt.ylabel("PR-AUC", fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(alpha=0.3)
    savefig(f"{model_name}_training_pr_auc.png")


def plot_confusion_matrix_highres(cm, model_name):
    labels = ["Non-Promoter", "Promoter"]
    plt.figure(figsize=(6, 5))
    plt.imshow(cm, interpolation="nearest")
    plt.title(f"{model_name}: Confusion Matrix", fontsize=14)
    plt.colorbar(label="Number of Sequences")
    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks, labels, rotation=20, ha="right", fontsize=10)
    plt.yticks(tick_marks, labels, fontsize=10)
    plt.xlabel("Predicted Label", fontsize=12)
    plt.ylabel("True Label", fontsize=12)

    thresh = cm.max() / 2.0 if cm.max() > 0 else 0
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(
                j,
                i,
                format(cm[i, j], "d"),
                ha="center",
                va="center",
                color="white" if cm[i, j] > thresh else "black",
                fontsize=13,
                fontweight="bold",
            )
    savefig(f"{model_name}_confusion_matrix.png")


def plot_roc_pr_curves(y_true, y_prob, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc_auc = roc_auc_score(y_true, y_prob)

    plt.figure(figsize=(7, 6))
    plt.plot(fpr, tpr, linewidth=2, label=f"ROC-AUC = {roc_auc:.4f}")
    plt.plot([0, 1], [0, 1], linestyle="--", linewidth=1.5, label="Random Classifier")
    plt.title(f"{model_name}: ROC Curve", fontsize=14)
    plt.xlabel("False Positive Rate", fontsize=12)
    plt.ylabel("True Positive Rate", fontsize=12)
    plt.legend(loc="lower right", fontsize=10)
    plt.grid(alpha=0.3)
    savefig(f"{model_name}_roc_curve.png")

    precision, recall, _ = precision_recall_curve(y_true, y_prob)
    pr_auc = average_precision_score(y_true, y_prob)

    plt.figure(figsize=(7, 6))
    plt.plot(recall, precision, linewidth=2, label=f"PR-AUC = {pr_auc:.4f}")
    plt.title(f"{model_name}: Precision-Recall Curve", fontsize=14)
    plt.xlabel("Recall", fontsize=12)
    plt.ylabel("Precision", fontsize=12)
    plt.legend(loc="lower left", fontsize=10)
    plt.grid(alpha=0.3)
    savefig(f"{model_name}_pr_curve.png")

# ============================================================
# 8. EVALUATION FUNCTION
# ============================================================


def evaluate_model(model, X, y_true, model_name):
    y_prob = model.predict(X, batch_size=BATCH_SIZE, verbose=0).ravel()
    y_pred = (y_prob >= 0.5).astype(int)

    metrics = {
        "Model": model_name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, zero_division=0),
        "Recall": recall_score(y_true, y_pred, zero_division=0),
        "F1": f1_score(y_true, y_pred, zero_division=0),
        "ROC_AUC": roc_auc_score(y_true, y_prob),
        "PR_AUC": average_precision_score(y_true, y_prob),
        "False_Positives": int(((y_true == 0) & (y_pred == 1)).sum()),
        "False_Negatives": int(((y_true == 1) & (y_pred == 0)).sum()),
    }
    cm = confusion_matrix(y_true, y_pred)
    return metrics, cm, y_prob, y_pred

# ============================================================
# 9. WEEK 12 HYPERPARAMETER TUNING
# ============================================================

print("\n" + "=" * 80)
print("WEEK 12: HYPERPARAMETER TUNING")
print("=" * 80)


tuning_space = [
    {"architecture": "CNN", "filters": 64, "kernel_size": 8, "dense_units": 128, "dropout": 0.50, "learning_rate": 1e-3},
    {"architecture": "CNN", "filters": 96, "kernel_size": 10, "dense_units": 128, "dropout": 0.50, "learning_rate": 1e-3},
    {"architecture": "Hybrid_CNN_BiLSTM", "filters": 64, "kernel_size": 8, "lstm_units": 64, "dense_units": 128, "dropout": 0.50, "learning_rate": 1e-3},
    {"architecture": "Hybrid_CNN_BiLSTM", "filters": 96, "kernel_size": 10, "lstm_units": 64, "dense_units": 128, "dropout": 0.50, "learning_rate": 1e-3},
]

early_stop_tuning = tf.keras.callbacks.EarlyStopping(
    monitor="val_roc_auc",
    mode="max",
    patience=2,
    restore_best_weights=True,
)

tuning_results = []

for i, hp in enumerate(tuning_space, start=1):
    tf.keras.backend.clear_session()
    print(f"\nTuning trial {i}/{len(tuning_space)}: {hp}")

    if hp["architecture"] == "CNN":
        model = build_cnn(
            input_shape=(MAX_LEN, 4),
            filters=hp["filters"],
            kernel_size=hp["kernel_size"],
            dense_units=hp["dense_units"],
            dropout=hp["dropout"],
            learning_rate=hp["learning_rate"],
        )
    else:
        model = build_hybrid_cnn_bilstm(
            input_shape=(MAX_LEN, 4),
            filters=hp["filters"],
            kernel_size=hp["kernel_size"],
            lstm_units=hp["lstm_units"],
            dense_units=hp["dense_units"],
            dropout=hp["dropout"],
            learning_rate=hp["learning_rate"],
        )

    history = model.fit(
        X_tr,
        y_tr,
        validation_data=(X_val, y_val),
        epochs=TUNING_EPOCHS,
        batch_size=BATCH_SIZE,
        class_weight=class_weights,
        callbacks=[early_stop_tuning],
        verbose=1,
    )

    val_prob = model.predict(X_val, batch_size=BATCH_SIZE, verbose=0).ravel()
    val_pred = (val_prob >= 0.5).astype(int)

    row = dict(hp)
    row.update({
        "val_accuracy": accuracy_score(y_val, val_pred),
        "val_precision": precision_score(y_val, val_pred, zero_division=0),
        "val_recall": recall_score(y_val, val_pred, zero_division=0),
        "val_f1": f1_score(y_val, val_pred, zero_division=0),
        "val_roc_auc": roc_auc_score(y_val, val_prob),
        "val_pr_auc": average_precision_score(y_val, val_prob),
        "epochs_completed": len(history.history["loss"]),
    })
    tuning_results.append(row)


tuning_df = pd.DataFrame(tuning_results).sort_values(
    by=["val_roc_auc", "val_pr_auc", "val_f1"], ascending=False
).reset_index(drop=True)

tuning_csv = OUTPUT_DIR / "week12_hyperparameter_tuning_results.csv"
tuning_df.to_csv(tuning_csv, index=False)
print("\nWeek 12 tuning results:")
print(tuning_df)
print("Saved:", tuning_csv)

best_hp = tuning_df.iloc[0].to_dict()

team_decision_text = f"""
Week 12 Hyperparameter Tuning Team Decision Log
================================================

we compared  CNN and Hybrid CNN-BiLSTM architectures using validation ROC-AUC as the primary selection metric, with PR-AUC and F1-score used as secondary checks. This keeps the model selection aligned with promoter classification, where ranking quality and class-specific error behavior are important.

Final selected architecture: {best_hp['architecture']}
Selected filters: {int(best_hp['filters'])}
Selected kernel size: {int(best_hp['kernel_size'])}
Selected dense units: {int(best_hp['dense_units'])}
Selected dropout: {float(best_hp['dropout'])}
Selected learning rate: {float(best_hp['learning_rate'])}
Selected LSTM units: {int(best_hp['lstm_units']) if not pd.isna(best_hp.get('lstm_units', np.nan)) else 'Not applicable'}

Team rationale:
1. The convolutional layer was kept because it acts as a motif detector over DNA sequences, similar to how regulatory sequence models learn local binding patterns.
2. The Hybrid CNN-BiLSTM option was tested because recurrent layers can capture long-range dependencies between motifs, following the biological motivation used in DanQ-style architectures.
3. Dropout, batch normalization, early stopping, and class weights were retained to reduce overfitting and maintain balanced performance.
4. The final architecture was selected collectively based on validation ROC-AUC first, then PR-AUC, F1-score, training stability, and biological interpretability.
"""

team_log_path = OUTPUT_DIR / "week12_team_architecture_decision_log.txt"
team_log_path.write_text(team_decision_text)
print("Saved:", team_log_path)
print(team_decision_text)

# ============================================================
# 10. FINAL TRAINING USING BEST PARAMETERS
# ============================================================

print("\n" + "=" * 80)
print("FINAL TRAINING")
print("=" * 80)

final_callbacks = [
    tf.keras.callbacks.EarlyStopping(
        monitor="val_roc_auc",
        mode="max",
        patience=4,
        restore_best_weights=True,
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=2,
        min_lr=1e-6,
        verbose=1,
    ),
]

final_model_specs = {
    "Final_CNN": {
        "architecture": "CNN",
        "filters": int(best_hp["filters"]),
        "kernel_size": int(best_hp["kernel_size"]),
        "dense_units": int(best_hp["dense_units"]),
        "dropout": float(best_hp["dropout"]),
        "learning_rate": float(best_hp["learning_rate"]),
    },
    "Final_Hybrid_CNN_BiLSTM": {
        "architecture": "Hybrid_CNN_BiLSTM",
        "filters": int(best_hp["filters"]),
        "kernel_size": int(best_hp["kernel_size"]),
        "lstm_units": int(best_hp["lstm_units"]) if not pd.isna(best_hp.get("lstm_units", np.nan)) else 64,
        "dense_units": int(best_hp["dense_units"]),
        "dropout": float(best_hp["dropout"]),
        "learning_rate": float(best_hp["learning_rate"]),
    },
}

final_results = []
saved_objects = {}

for model_name, hp in final_model_specs.items():
    tf.keras.backend.clear_session()
    print("\n" + "=" * 70)
    print(f"Training {model_name}")
    print("=" * 70)
    print(hp)

    if hp["architecture"] == "CNN":
        model = build_cnn(
            input_shape=(MAX_LEN, 4),
            filters=hp["filters"],
            kernel_size=hp["kernel_size"],
            dense_units=hp["dense_units"],
            dropout=hp["dropout"],
            learning_rate=hp["learning_rate"],
        )
    else:
        model = build_hybrid_cnn_bilstm(
            input_shape=(MAX_LEN, 4),
            filters=hp["filters"],
            kernel_size=hp["kernel_size"],
            lstm_units=hp["lstm_units"],
            dense_units=hp["dense_units"],
            dropout=hp["dropout"],
            learning_rate=hp["learning_rate"],
        )

    model.summary()

    history = model.fit(
        X_tr,
        y_tr,
        validation_data=(X_val, y_val),
        epochs=FINAL_EPOCHS,
        batch_size=BATCH_SIZE,
        class_weight=class_weights,
        callbacks=final_callbacks,
        verbose=1,
    )

    plot_training_history(history, model_name)

    metrics, cm, y_prob, y_pred = evaluate_model(model, X_test, y_test, model_name)
    final_results.append(metrics)

    print(f"\nClassification Report: {model_name}")
    print(classification_report(y_test, y_pred, target_names=["Non-Promoter", "Promoter"]))
    print("Confusion Matrix:\n", cm)
    print("Metrics:", metrics)

    plot_confusion_matrix_highres(cm, model_name)
    plot_roc_pr_curves(y_test, y_prob, model_name)

    model_path = OUTPUT_DIR / f"{model_name}.keras"
    model.save(model_path)
    print("Saved model:", model_path)

    saved_objects[model_name] = {
        "model": model,
        "history": history,
        "metrics": metrics,
        "cm": cm,
        "y_prob": y_prob,
        "y_pred": y_pred,
        "model_path": str(model_path),
    }

results_df = pd.DataFrame(final_results).sort_values(by="ROC_AUC", ascending=False).reset_index(drop=True)
results_path = OUTPUT_DIR / "final_model_comparison.csv"
results_df.to_csv(results_path, index=False)

print("\n" + "=" * 80)
print("FINAL MODEL COMPARISON")
print("=" * 80)
print(results_df)
print("Saved:", results_path)

best_model_name = results_df.loc[0, "Model"]
best_model = saved_objects[best_model_name]["model"]
best_prob = saved_objects[best_model_name]["y_prob"]
best_pred = saved_objects[best_model_name]["y_pred"]

print(f"\nBest final model based on ROC-AUC: {best_model_name}")

# ============================================================
# 11. DANQ BENCHMARK COMPARISON
# ============================================================

benchmark_df = results_df.copy()
benchmark_df["DanQ_Reference_ROC_AUC"] = DANQ_REFERENCE_ROC_AUC
benchmark_df["ROC_AUC_minus_DanQ_Reference"] = benchmark_df["ROC_AUC"] - DANQ_REFERENCE_ROC_AUC
benchmark_df["DanQ_Note"] = DANQ_REFERENCE_NOTE
benchmark_path = OUTPUT_DIR / "danq_benchmark_comparison.csv"
benchmark_df.to_csv(benchmark_path, index=False)

print("\nDanQ benchmark comparison:")
print(benchmark_df[["Model", "ROC_AUC", "DanQ_Reference_ROC_AUC", "ROC_AUC_minus_DanQ_Reference"]])
print("Saved:", benchmark_path)

# ============================================================
# 12. FALSE POSITIVE INVESTIGATION
# ============================================================

print("\n" + "=" * 80)
print("FALSE POSITIVE MOTIF INVESTIGATION")
print("=" * 80)

known_promoter_motifs = {
    "TATA_box": "TATAAA",
    "CAAT_box": "CCAAT",
    "GC_box": "GGGCGG",
    "Inr_core": "YYANWYY",
}


def contains_iupac_pattern(seq, pattern):
    iupac = {
        "A": "A", "C": "C", "G": "G", "T": "T",
        "Y": "[CT]", "R": "[AG]", "W": "[AT]", "S": "[CG]",
        "M": "[AC]", "K": "[GT]", "N": "[ACGT]",
    }
    regex = "".join(iupac.get(ch, ch) for ch in pattern)
    return re.search(regex, seq) is not None


def count_known_motifs(seq):
    seq = clean_sequence(seq)
    return {
        "TATA_box_count": seq.count("TATAAA"),
        "CAAT_box_count": seq.count("CCAAT"),
        "GC_box_count": seq.count("GGGCGG"),
        "Inr_core_present": int(contains_iupac_pattern(seq, "YYANWYY")),
    }


def kmer_counts(seqs, k=6):
    counter = Counter()
    for seq in seqs:
        seq = clean_sequence(seq)
        for i in range(0, max(0, len(seq) - k + 1)):
            kmer = seq[i:i+k]
            if "N" not in kmer:
                counter[kmer] += 1
    return counter


def top_enriched_kmers(group_a_seqs, group_b_seqs, k=6, top_n=25, pseudocount=1.0):
    a_counts = kmer_counts(group_a_seqs, k=k)
    b_counts = kmer_counts(group_b_seqs, k=k)
    all_kmers = set(a_counts) | set(b_counts)
    total_a = sum(a_counts.values()) + pseudocount * len(all_kmers)
    total_b = sum(b_counts.values()) + pseudocount * len(all_kmers)

    rows = []
    for kmer in all_kmers:
        freq_a = (a_counts[kmer] + pseudocount) / total_a
        freq_b = (b_counts[kmer] + pseudocount) / total_b
        rows.append({
            "kmer": kmer,
            "false_positive_frequency": freq_a,
            "true_negative_frequency": freq_b,
            "enrichment_ratio_FP_vs_TN": freq_a / freq_b,
            "false_positive_count": a_counts[kmer],
            "true_negative_count": b_counts[kmer],
        })

    return pd.DataFrame(rows).sort_values(
        by="enrichment_ratio_FP_vs_TN", ascending=False
    ).head(top_n)


false_positive_idx = np.where((y_test == 0) & (best_pred == 1))[0]
true_negative_idx = np.where((y_test == 0) & (best_pred == 0))[0]
true_positive_idx = np.where((y_test == 1) & (best_pred == 1))[0]
false_negative_idx = np.where((y_test == 1) & (best_pred == 0))[0]

fp_sequences = [test_sequences[i] for i in false_positive_idx]
tn_sequences = [test_sequences[i] for i in true_negative_idx]
tp_sequences = [test_sequences[i] for i in true_positive_idx]
fn_sequences = [test_sequences[i] for i in false_negative_idx]

print(f"False positives: {len(fp_sequences)}")
print(f"True negatives:  {len(tn_sequences)}")
print(f"True positives:  {len(tp_sequences)}")
print(f"False negatives: {len(fn_sequences)}")

fp_motif_rows = []
for idx in false_positive_idx:
    seq = test_sequences[idx]
    row = {
        "test_index": int(idx),
        "true_label": "Non-Promoter",
        "predicted_label": "Promoter",
        "predicted_promoter_probability": float(best_prob[idx]),
        "sequence_start": seq[:80],
        "sequence_full": seq,
    }
    row.update(count_known_motifs(seq))
    fp_motif_rows.append(row)

fp_motif_df = pd.DataFrame(fp_motif_rows).sort_values(
    by="predicted_promoter_probability", ascending=False
)
fp_motif_path = OUTPUT_DIR / "false_positive_sequences_with_promoter_motif_counts.csv"
fp_motif_df.to_csv(fp_motif_path, index=False)
print("Saved:", fp_motif_path)

if len(fp_sequences) > 0 and len(tn_sequences) > 0:
    enriched_df = top_enriched_kmers(fp_sequences, tn_sequences, k=6, top_n=30)
    enriched_path = OUTPUT_DIR / "false_positive_enriched_6mers_vs_true_negatives.csv"
    enriched_df.to_csv(enriched_path, index=False)
    print("\nTop 6-mers enriched in false positives compared with true negatives:")
    print(enriched_df)
    print("Saved:", enriched_path)
else:
    enriched_df = pd.DataFrame()
    print("Not enough false positives or true negatives for k-mer enrichment analysis.")

# Summary motif frequency table across FP, TN, TP, FN
error_groups = {
    "False_Positive_NonPromoter_PredPromoter": fp_sequences,
    "True_Negative_NonPromoter_PredNonPromoter": tn_sequences,
    "True_Positive_Promoter_PredPromoter": tp_sequences,
    "False_Negative_Promoter_PredNonPromoter": fn_sequences,
}

motif_summary_rows = []
for group_name, seqs in error_groups.items():
    if len(seqs) == 0:
        motif_summary_rows.append({"Group": group_name, "N": 0})
        continue
    motif_counts = [count_known_motifs(s) for s in seqs]
    df = pd.DataFrame(motif_counts)
    row = {"Group": group_name, "N": len(seqs)}
    for col in df.columns:
        row[f"mean_{col}"] = df[col].mean()
        row[f"sequences_with_{col}"] = int((df[col] > 0).sum())
        row[f"percent_with_{col}"] = float((df[col] > 0).mean() * 100)
    motif_summary_rows.append(row)

motif_summary_df = pd.DataFrame(motif_summary_rows)
motif_summary_path = OUTPUT_DIR / "motif_summary_by_confusion_matrix_group.csv"
motif_summary_df.to_csv(motif_summary_path, index=False)
print("\nKnown promoter motif summary by confusion-matrix group:")
print(motif_summary_df)
print("Saved:", motif_summary_path)

# ============================================================
# 13. BIOLOGICAL INTERPRETATION FOR FINAL REPORT
# ============================================================

best_metrics = results_df.iloc[0].to_dict()

biological_interpretation = f"""
Final Report Biological Interpretation
======================================

Best model: {best_model_name}
Test ROC-AUC: {best_metrics['ROC_AUC']:.4f}
Test PR-AUC: {best_metrics['PR_AUC']:.4f}
Test F1-score: {best_metrics['F1']:.4f}

Connection to regulatory regions:
Promoters are regulatory DNA regions that help control transcription initiation. In this project, the CNN portion of the model acts as an automatic feature extractor over one-hot encoded DNA. The convolutional filters scan across each sequence and learn short local patterns that may resemble promoter-associated motifs, such as TATA-box-like, CAAT-box-like, GC-rich, or initiator-like signals. These learned filters are conceptually similar to motif detectors used in regulatory genomics models.

Connection to DeepBind:
The CNN feature extraction step follows the same biological idea used by DeepBind-style models: convolutional filters can learn sequence patterns even when the exact motif location varies across sequences. This is important for regulatory-region prediction because transcription-factor binding sites and promoter motifs do not always appear at a fixed position.

Connection to DanQ:
The Hybrid CNN-BiLSTM model extends the CNN by adding a bidirectional recurrent layer. The CNN captures local motifs, while the BiLSTM captures dependencies between motifs across longer sequence ranges. This follows the DanQ-style biological motivation that regulatory activity depends not only on individual motifs but also on motif combinations, spacing, and longer-range regulatory grammar.

False-positive interpretation:
False positives are non-promoter sequences predicted as promoters. These cases are biologically important because some non-promoter regions may contain promoter-like motifs or GC-rich regulatory patterns. The saved false-positive motif tables and enriched k-mer table should be used in the final report to discuss whether the model is being misled by non-promoter sequences that share promoter-associated sequence features.

DanQ benchmark interpretation:
The DanQ benchmark comparison table compares this project's ROC-AUC against a configurable DanQ reference value. Because DanQ reports performance over regulatory genomic targets rather than this exact promoter-only dataset, the comparison should be presented carefully as a literature reference rather than a perfectly identical dataset comparison.

Professional figures:
All ROC curves, PR curves, training curves, and confusion matrices are saved as high-resolution 300-DPI PNG files in the promoter_final_outputs folder.
"""

bio_path = OUTPUT_DIR / "final_report_biological_interpretation.txt"
bio_path.write_text(biological_interpretation)
print("\n" + biological_interpretation)
print("Saved:", bio_path)

# ============================================================
# 14. SINGLE-SEQUENCE PREDICTION FUNCTION
# ============================================================


def predict_sequence(seq, model=best_model, max_len=MAX_LEN):
    x = one_hot_encode_sequence(seq, max_len=max_len)
    x = np.expand_dims(x, axis=0)
    prob = float(model.predict(x, verbose=0)[0][0])
    label = "Promoter" if prob >= 0.5 else "Non-Promoter"
    return {
        "sequence_start": seq[:80],
        "predicted_label": label,
        "probability_promoter": prob,
    }

print("\nExample prediction using best model:")
print(predict_sequence(test_sequences[0]))

print("\nAll outputs saved in:", OUTPUT_DIR.resolve())
```
